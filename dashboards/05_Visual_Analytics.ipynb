# E-Commerce Analytics Dashboard
*Built on Gold Layer Views*

## Executive

```sql
/*
======================================================
Goal: High-level business health for the CEO.
======================================================
*/
-- 1. Tell Databricks which schema to use
-- (Replace 'default' with your schema name if you created one, e.g., 'analysis')
USE analysis;

-- KPI 1: Monthly Revenue Growth
SELECT month, revenue, total_orders 
FROM analysis_monthly_revenue;
```

```sql
-- KPI 2: Category Market Share
SELECT category_en, SUM(total_item_value) as total_sales
FROM gold.fact_order_items oi
JOIN gold.dim_products p ON oi.product_id = p.product_id
GROUP BY 1 ORDER BY 2 DESC LIMIT 10;
```

## Logistics

```sql
/*
======================================================
Goal: Identify bottlenecks for the Operations Manager.
======================================================
*/
-- KPI 1: Delivery Delay Heatmap
SELECT customer_state, avg_delivery_days, delay_rate 
FROM analysis.delivery_performance_by_state;
```

```sql
-- KPI 2: Estimated vs. Actual Delivery Time
SELECT 
    date_trunc('week', order_purchase_timestamp) as week,
    AVG(actual_delivery_days) as avg_actual,
    AVG(estimated_delivery_days) as avg_estimated
FROM gold.fact_orders
GROUP BY 1 ORDER BY 1;
```

## Customer

```sql
/*
======================================================
Goal: Understand buyer behavior for the Marketing Manager.
======================================================
*/
-- KPI 1: Customer Concentration: How many customers have 1 order vs 2+ orders?
SELECT 
    CASE WHEN total_orders = 1 THEN 'Single Buyer' ELSE 'Repeat Buyer' END as customer_type,
    COUNT(customer_id) as customer_count
FROM analysis_customer_metrics
GROUP BY 1;
```

```sql
-- KPI 2: Geography of Sales
SELECT customer_city, COUNT(customer_id) as order_count
FROM gold.dim_customers
GROUP BY 1 ORDER BY 2 DESC LIMIT 20;
```

## Payment

```sql
/*
======================================================
Goal: Finance and risk analysis.
======================================================
*/
-- KPI 1: Payment Method Popularity
SELECT payment_type, COUNT(*) as usage_count, SUM(payment_value) as total_value
FROM gold.fact_payments
GROUP BY 1 ORDER BY 3 DESC;
```

```sql
-- KPI 2: Installment vs. Order Value
-- Do expensive items lead to more installments?
SELECT payment_installments, AVG(payment_value) as avg_value
FROM gold.fact_payments
WHERE payment_installments > 0
GROUP BY 1 ORDER BY 1;
```
