```sql
/*
======================================================
Goal: High-level business health for the CEO.
======================================================
*/

-- KPI 1: Monthly Revenue Growth
SELECT month, revenue, total_orders 
FROM analysis_monthly_revenue;

-- KPI 2: Category Market Share
SELECT category_en, SUM(total_item_value) as total_sales
FROM gold.fact_order_items oi
JOIN gold.dim_products p ON oi.product_id = p.product_id
GROUP BY 1 ORDER BY 2 DESC LIMIT 10;
```
