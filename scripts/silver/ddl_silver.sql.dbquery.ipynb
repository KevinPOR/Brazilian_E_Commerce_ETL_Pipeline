```sql
CREATE TABLE silver.orders
LIKE bronze.orders;

CREATE TABLE silver.customers
LIKE bronze.customers;

CREATE TABLE silver.geolocation
LIKE bronze.geolocation;

CREATE TABLE silver.order_items 
AS 
SELECT 
    *, 
    CAST(0 AS DOUBLE) AS total_item_value, 
    CAST(0 AS DOUBLE) AS shipping_ratio 
FROM bronze.order_items 
WHERE 1=0; -- This ensures we only copy the schema, not the data

CREATE TABLE silver.order_payments 
AS 
SELECT 
    *,
    CAST(0 AS DOUBLE) AS installment_value 
FROM bronze.order_payments 
WHERE 1=0; -- This ensures we only copy the schema, not the data

CREATE TABLE silver.order_reviews
LIKE bronze.order_reviews;

CREATE TABLE silver.product_category_name_translation
LIKE bronze.product_category_name_translation;

CREATE TABLE silver.products
LIKE bronze.products;

CREATE TABLE silver.sellers
LIKE bronze.sellers;
```
