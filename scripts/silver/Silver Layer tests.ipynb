##order Table tests

```sql
-- Status normalization works
SELECT DISTINCT order_status, TRIM(order_status)
FROM silver.orders
WHERE order_status != TRIM(order_status);

-- Required columns not null
SELECT *
FROM silver.orders
WHERE order_id IS NULL
   OR order_purchase_timestamp IS NULL;

-- Test your transformation logic
SELECT *
FROM silver.orders
WHERE order_approved_at IS NOT NULL
  AND order_approved_at <= order_purchase_timestamp;

SELECT *
FROM silver.orders
WHERE order_delivered_carrier_date IS NOT NULL
  AND order_delivered_carrier_date <= order_approved_at;


SELECT *
FROM silver.orders
WHERE order_status = 'delivered'
  AND order_delivered_customer_date IS NULL;

SELECT *
FROM silver.orders
WHERE order_delivered_customer_date IS NOT NULL
  AND order_delivered_customer_date < order_delivered_carrier_date;

SELECT *
FROM silver.orders
WHERE order_status != 'delivered'
  AND order_delivered_customer_date IS NOT NULL;

SELECT order_id, COUNT(*)
FROM silver.orders
GROUP BY order_id
HAVING COUNT(*) > 1;

SELECT *
FROM silver.orders
WHERE NOT (
  order_purchase_timestamp <= order_approved_at
  AND order_approved_at <= order_delivered_carrier_date
  AND order_delivered_carrier_date <= order_delivered_customer_date
);
```
