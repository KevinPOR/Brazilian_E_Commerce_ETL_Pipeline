##Risk dashboard

##Customer risk view

```sql
SELECT
  customer_id,
  AVG(prediction) AS risk_score
FROM analysis.predicted_late_deliveries
GROUP BY customer_id;
```

```sql
SELECT
    COUNT(*) AS total_orders,
    AVG(prediction) AS late_rate
FROM analysis.predicted_late_deliveries;
```

```sql
-- KPI 2: Late Deliveries by Customer
SELECT
    customer_id,
    AVG(prediction) AS risk_score
FROM analysis.predicted_late_deliveries
GROUP BY customer_id
ORDER BY risk_score DESC;
```

```sql
-- KPI 3: Payment Type Risk
SELECT
    payment_type,
    AVG(prediction) AS late_probability
FROM analysis.predicted_late_deliveries
JOIN gold.fact_payments USING(order_id)
GROUP BY payment_type;
```
